# Release Year

```python
previous_decade = df[["type", "release_year"]]
previous_decade = previous_decade.rename(columns = {"release_year" : "Release Year"})
previous_decade = previous_decade[previous_decade["Release Year"]>=2010]
previous_decade
```

```python
previous_decade_df = previous_decade.groupby("Release Year")["type"].size().reset_index()
previous_decade_df = pd.DataFrame(previous_decade_df)
previous_decade_df
```

```python
previous_decade_df.rename(columns = {"type": "Total Content"}, inplace = True)
```

```python
previous_decade.groupby("Release Year")["type"].value_counts()
```

```python
plt.figure(figsize = (10,6))
count_plot = sns.countplot(x = "Release Year", data = previous_decade, hue="type",
             palette= "PuBuGn")
count_plot.set(title = "Trend of each type of content Released over the years")
```

```python
plt.figure(figsize = (10,6))
plot_total_content= sns.lineplot(x= "Release Year", y = "Total Content", data = previous_decade_df,
                                linewidth = 3)
plot_total_content.set(xlabel = "Release Year", ylabel = "Total Content",
                      title = "Trend of content on netflix")
plt.show()
```

Result: It is seen that most content on netflix was released in year 2018, the amount of content is decreasing since then.

# Rating

```python
df.rating.unique()
```

```python
new_catgs = {
    'TV-PG': 'Parental Guidance',
    'TV-MA' : 'Mature Audience',
    'TV-Y7-FV': 'Teens',
    'TV-Y7': 'Teens',
    'TV-14': 'Teens',
    'R': 'Mature Audience',
    'TV-Y': 'General Audience',
    'NR': 'Mature Audience',
    'PG-13': 'Teens',
    'TV-G': 'General Audience',
    'PG': 'Teens',
    'G': 'General Audience',
    'UR': 'Mature Audience',
    'NC-17': 'Mature Audience'
}
df['rating']=df['rating'].replace(new_catgs)
df.head()
```

```python
plt.figure(figsize= (10,6))
sns.countplot(x="rating", data=df, palette="PuBuGn",)
plt.title("count of Rating by Movie and Shows");
```

Result: It is seen that Netflix has more content for Mature audience.
